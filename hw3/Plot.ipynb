```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
```

```python
acc_loss = np.loadtxt("loss_acc.csv", dtype=np.str, delimiter=",")
```

```python
loss = []
acc = []
for item in acc_loss:
    loss.append(item[0][-8:])
    acc.append(item[1][-8:])
loss = np.array(loss,dtype =float)
acc = np.array(acc,dtype =float)
```

```python
plt.plot(acc)
plt.show()
plt.plot(loss)
plt.show
```
