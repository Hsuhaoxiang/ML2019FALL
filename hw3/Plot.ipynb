```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
```

```python
acc_loss = np.loadtxt("loss_acc.csv", dtype=np.str, delimiter=",")
```

```python
t_loss = []
t_acc = []
v_loss = []
v_acc = []
print(acc_loss)
for i in range(int(len(acc_loss)/2)):
    t_loss.append(float(acc_loss[2*i][1]))
    v_loss.append(float(acc_loss[2*i+1][1]))
    t_acc.append(float(acc_loss[2*i][2]))
    v_acc.append(float(acc_loss[2*i+1][2]))
axis = [i for i in range(600)]
```

```python
plt.plot(axis,t_loss,axis,v_loss)
plt.show()
plt.plot(axis,t_acc,axis,v_acc)
plt.show
```
